# Sample Sentiment Classifier Submission
This notebook contains a sample heuristic model to classify movie review sentiments (positive/negative).

```python
# Cell 1: Import core dependencies
import os
import math
```

```python
# Cell 2: Heuristic Sentiment Predictor Function
def predict(inputs):
    """
    Receives a list of string movie reviews and returns prediction labels:
    1 for Positive sentiment, 0 for Negative sentiment.
    """
    positives = {'good', 'great', 'love', 'amazing', 'excellent', 'wonderful', 'like', 'best', 'happy'}
    negatives = {'bad', 'worst', 'hate', 'boring', 'waste', 'terrible', 'awful', 'poor', 'no'}
    
    preds = []
    for text in inputs:
        words = text.lower().split()
        p_count = sum(1 for w in words if w in positives)
        n_count = sum(1 for w in words if w in negatives)
        
        # Classify positive if positive terms outweigh negative ones
        preds.append(1 if p_count >= n_count else 0)
        
    return preds
```
